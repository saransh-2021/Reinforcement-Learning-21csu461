```python
import numpy as np
import plotly.express as px
```

```python
np.random.seed(42)
```

```python
# 1000 random integers between 0 and 50
x = np.random.randint(0, 50, 1000)
x
```

```python
# Positive correlation with some noise
y = x + np.random.normal(0, 10, 1000)
y
```

```python
np.corrcoef(x,y)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

matplotlib.style.use('ggplot')

plt.scatter(x,y)
plt.show()
#positive
```

```python
px.scatter(x= x, y= y)
```

```python
y = x - np.random.normal(0, 10, 1000)
y
```

```python
np.corrcoef(x,y)
```

```python
plt.scatter(x, y)
plt.show()
#positive
```

```python
px.scatter(x= x, y= y)
```

```python
x = np.random.randint(0, 50, 1000)

y= 100 - x + np.random.normal(0, 5, 1000)

np.corrcoef(x, y)
```

```python
plt.scatter(x,y)
```

```python
px.scatter(x=x,y=y)
```

```python
x = np.random.randint(0, 50, 1000)
y = x + np.random.normal(0, 25, 1000)

z = 100 - x + np.random.normal(0,25,1000)
```

```python
np.corrcoef(x,y) , np.corrcoef(x,z)
```

```python
plt.scatter(x, y)
plt.scatter(x, z)
```

```python
x = np.random.randint(0,50,1000)
y = np.random.randint(0,50,1000)

print(np.corrcoef)

plt.scatter(x, y)
```

```python
import pandas as pd

df = pd.DataFrame({'a':np.random.randint(0,50,1000)})
df['b'] = df['a'] + np.random.normal(0,10,1000)
df['c'] = 100 - df['a'] + np.random.normal(0,10,1000)
df['d'] = np.random.randint(0,50,1000)

df.corr()
```

```python
pd.plotting.scatter_matrix(df, figsize = (6,6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)), df.columns)
plt.yticks(range(len(df.columns)), df.columns)
plt.colorbar()
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(), annot=True)
```
